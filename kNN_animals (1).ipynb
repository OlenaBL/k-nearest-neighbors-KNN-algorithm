```python
# First Feature
species = ['Frog','Panda','Cod','LeopardFrog', 'Dolphin', 'Clownfish']
# Second Feature

breathing_env = ['Both', 'Air','Water', 'Both','Air','Water']

# Label
species_type = ['Amphibian', 'Mammal', 'Fish', 'Amphibian', 'Mammal', 'Fish']
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()

species_encoded=le.fit_transform(species)
print(species_encoded)
```

```python
breathing_env_encoded=le.fit_transform(breathing_env)
label=le.fit_transform(species_type)
features=list(zip(species_encoded, breathing_env_encoded))
print(features)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=3)

# Training
model.fit(features, label)
#Predictions
predicted = model.predict([[3,2]])
predicted_print = le.inverse_transform(predicted) #converts from numerical value
print(predicted_print)
```
